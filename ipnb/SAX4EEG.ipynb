# First studies of EEG signals 
In order to do some initial analysis of EEG signals we download the UCI [EEG Eye State Data Set](https://archive.ics.uci.edu/ml/datasets/EEG+Eye+State#). 

+ Dataset obtained from a continuous EEG measurement.
+ EEG technology used: `Emotiv EEG Neuroheadset`. 
+ Duration: 117 seconds. 
+ Last column: Eye state detected using a camera and added manually. (1: eye-closed, 0: eye-opened)
+ Values in chronological order (from the top to the bottom of the dataset)
+ Since the dataset contains 14980 points, the frequency of measures is of 128 points per second, which is close to 10 times the [frequency of Beta wave](http://emedicine.medscape.com/article/1139332-overview#a2). 


# Filtering outliers
The EEG records contain some unusually high values. It is better to filter these measures in order to have a better repartition of the data points after the z-score standardization

```python
def clean_outlier(column):
    mean_column = column.mean()
    std_column = column.std()
    column[column > mean_column+3*std_column] = mean_column+3*std_column
    column[column < mean_column-3*std_column] = mean_column-3*std_column
    return column
```

# Load the dataset
+ Use the Pandas library to read the csv file 
+ Separate the records and the labels

```python
eeg = pd.read_csv("../data/EEG_Eye_State.arff",comment="@",header=None)
Y = eeg[eeg.columns[-1]]
X = eeg.drop(eeg.columns[-1],1)
X = X.apply(clean_outlier)
```

# Standardize the dataset 

```python
eeg_id = 2
z_eeg = znormalization(X)
plt.plot(range(Y.size),z_eeg[eeg_id],alpha=0.5)
plt.plot(range(Y.size),Y)
plt.ylim([-3,3])
```

# Compute the Piecewise Aggregate Approximation (PAA)
The PAA algorithm applies to the time series the following operations:
    + Segments the time series into equal parts 
    + Stores the average value for each part

```python
eeg_id = 2
split_eeg = paa_transform(z_eeg, 1500)
plt.plot(range(split_eeg[:,eeg_id].size),split_eeg[:,eeg_id])
plt.ylim([-3,3])
```

# Compute the Symbolic Aggregate approXimation (SAX)
The SAX algorithm applies to the time series the following operations:
+ Applies the PAA algorithm 
+ Converts the PAA results into symbols 

```python
sax_eeg = sax_transform(X, 1500, 30, False)
```

# From PAA to SAX representation
+ In red the discretization boundaries obtained when the normal distribution hypothesis is used
+ In black when the percentiles of the points distributions are used as boundaries

```python
alphabet_sz=30
eeg_id = 2
percentils2 = np.percentile(split_eeg,np.linspace(1./alphabet_sz, 
                                          1-1./alphabet_sz, 
                                          alphabet_sz-1)*100)

percentils = norm.ppf(np.linspace(1./alphabet_sz, 
                                          1-1./alphabet_sz, 
                                          alphabet_sz-1))
plt.plot(range(split_eeg[:,eeg_id].size),split_eeg[:,eeg_id],alpha=0.7)
for percentil in percentils:
    plt.plot((0, split_eeg[:,eeg_id].size), (percentil, percentil), 'k--')
for percentil in percentils2:
    plt.plot((0, split_eeg[:,eeg_id].size), (percentil, percentil), 'r:')
plt.ylim([-2.5,2.5])
```

# Graphic representation of the SAX transformation
+ Each symbol is denoted by a specific color, a grandient of colors has been chosen since the different symbolos could be considered as ordered


```python
palette = sns.color_palette("coolwarm", 30)
sns.palplot(sns.color_palette("coolwarm", 30))
```

# Plot of the SAX representation of the EEG
+ Each one of the 14 EEG records is converted to its SAX representation
+ The palette defined previously is used to represent the symbol at each time step
+ Notice that the different EEG records are more or less correlated

```python
step = 1./sax_eeg.shape[1]
for k,eeg_id in enumerate(range(sax_eeg.shape[1])):
    for i,val in enumerate(sax_eeg[:,eeg_id]):
        plt.plot((i,i), (k*step,(k+1)*step), color=palette[val])
```
